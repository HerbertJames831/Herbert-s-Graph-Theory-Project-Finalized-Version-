# Conditions for Graph Isomorphism

#### We are able to ascertain that an isomorphism exists between two graphs if the following conditions are satisfied

#### 1. An equal amount of edges in both given graphs

#### 2. An equal quantity of nodes in both graphs

#### 3. The same amount of degree sequence in both graphs

##### 4. There must be an identical cycle length in both graphs

# What is the Graph Isomorphism Problem?


#### The computational problem of ascertaining whether two finite graphs are isomorphic

#### The graph isomorphism is not generally known to be NP-complete or to be solved in polynomial time

#### However the graph isomorphism problem is widely known to be in the low hierarchy of the NP class; this demonstrates that unless the polynomial time hierarchy breaks down to its second level; it is not NP-complete

#### There is a possibility the problem might be NP-intermediate; NP-intermediate is a computational complexity class

# What is the Subgraph Isomorphism Problem ?

#### A computational task in which two graphs G and H are provided 

#### It must be ascertained whether G has a subgraph that is isomorphic to H

#### Subgraph Isomorphism must be formulated as a decision problem to prove it is NP-complete

#### For example a pair of graphs G and H is the input to the decision problem

#### If H is isomorphic to a subgraph of G, the answer to the problem is positive

#### Otherwise if H is not not isomorphic to a subgraph of G, the answer to the problem is negative

#### The subgraph isomorphism problem is generally known to be NP-complete

#### The graph isomorphism problem is known to be a special case of the subgraph isomorphism problem
